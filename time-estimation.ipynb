```python
def is_weekday(date):
    return (date.weekday() in [0,1,2,3,4])
```

```python
def went_over_night(start_datetime, end_datetime, new_datetime):
    return start_datetime < new_datetime and new_datetime < end_datetime

def elapsed_without_non_work_hours(start_dt, end_dt):
    new_hour = 23
    new_dt = datetime.datetime(start_dt.year,
                                     start_dt.month,
                                         start_dt.day,
                                         new_hour,
                                         start_dt.minute,
                                         start_dt.second)
    
    
    elapsed = end_dt - start_dt
    elapsed_days = 0
    new_dt = datetime.datetime(start_dt.year,
                                     start_dt.month,
                                         start_dt.day,
                                         new_hour,
                                         start_dt.minute,
                                         start_dt.second)
    while(went_over_night(start_dt, end_dt, new_dt)):
        if (is_weekday(new_dt)):
            elapsed -= datetime.timedelta(hours=16)
        else:
            elapsed -= datetime.timedelta(hours=24)
        elapsed_days += 1
        
        new_dt = new_dt + pd.DateOffset(months=0, days=1)

    return elapsed 
            
        
```

```python
def remove_extra_zeros(vector):
    rows_len = len(vector)
    for y in range(rows_len-1, 0, -1):
        if (vector[y] > 0):
            return vector[0:y+1]
    return vector

```

```python
def create_absolute_freq_vector(df):
    vec = pd.Series([])
    for x in range(0,100):
        abs_freq = len(df[df['rounded_elapsed'] == x])
        vec = vec.set_value(x, abs_freq)
    return vec
```

```python
def get_relative_freq_vector(series):
    summation = series.sum()
    return series / float(summation)
    
```

```python
def plot_vec(vec):
    pd.DataFrame({ 'estimate': pd.Series(vec)}).plot()
    plt.xlabel('time elapsed (hours)')
    plt.ylabel('probability')


plot_vec(estimates(5,0))
```

```python
def estimates(num_ones_estimates, num_twos_estimates):
    ones_rel_convolved = [1]

    for x in range(num_ones_estimates):
        ones_rel_convolved = np.convolve(ones_rel_convolved, ones_rel_vec)

    twos_rel_convolved = [1]    
    for y in range(num_twos_estimates):
        twos_rel_convolved = np.convolve(twos_rel_convolved, twos_rel_vec)

    return np.convolve(ones_rel_convolved, twos_rel_convolved)

def prob_get_done(num_ones_estimates, num_twos_estimates, hours):
    if (num_ones_estimates == 0 and num_twos_estimates == 0):
        return 0.0;
    est = estimates(num_ones_estimates, num_twos_estimates)
    range_of_hours = np.arange(len(est))
    plot_vec(est)
    estimate = est[0:hours].sum()
    plt.fill_between(range_of_hours, est, color = 'blue', where = (range_of_hours >=0) & (range_of_hours <= hours), alpha=0.5)
    plt.title("p(" + str(num_ones_estimates) + " 1-pt & " + str(num_twos_estimates) + " 2-pts under " + str(hours) + " hrs) = " + str(round(estimate, 3)))
    return estimate
```
